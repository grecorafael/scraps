```python
### EXERCÍCIO 4 - AUTOMATIZAÇÃO DE BUSCA (PREVIÃO DO TEMPO)

from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import Select
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
import requests


CidadesBusca = ['Bauru, São Paulo, Brasil', 'Marília, São Paulo, Brasil', 'Ocauçu, São Paulo, Brasil']

#Browser Visible
browser = webdriver.Chrome(executable_path="C:/AULA_DATA_SCIENCE/chromedriver")
url = 'https://weather.com/pt-BR/clima/10dias/l/BRMG0645:1:BR';
browser.get(url)

for cidade in CidadesBusca:
    txtName = cidade + ".txt";
    f = open(txtName,"w+")
    print("Cidade escolhida: ", cidade)
    fieldSearch = driver.find_element(By.CSS_SELECTOR, ".theme__inputElement__4bZUj.input__inputElement__1GjGE")
    time.sleep(1);
    fieldSearch.send_keys(cidade)
    time.sleep(1)
    fieldClicked = driver.find_element(By.CSS_SELECTOR, ".styles__item__3sdr8.styles__selected__SEH0e")
    time.sleep(1)
    fieldClicked.click()
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    table = soup.find('table', attrs={'class':'twc-table'})
    data = []
    table_body = table.find('tbody')
    rows = table_body.find_all('tr')
    for row in rows:
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        data.append([ele for ele in cols if ele])
        info = [ele for ele in cols if ele]
        post = {
         "Dia" : info[0],
         "Descrição" : info[1],
         "Max/Min" : info[2],
         "Precipitação" : info[3],
         "Vento" : info[4],
         "Umidade" : info[5]
         }
        txtLine = "Dia: " + info[0] + " Descrição: " + info[1] + " Max/Min: " + info[2] + " Precipitação: " + info[3] + " Vento: " + info[4] + " Umidade: " + info[5]
        txtLine += "\n"
        f.write(txtLine)
        f.write("---------------\n")
        print(post)    
        print("---------------")
        
browser.close()

```
